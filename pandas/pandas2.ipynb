```python
#distance en pc des planetes découvertes par année
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 20)
df9 = pd.read_csv("data20230518.csv")
df9["couleurs"]= "black"
df9["dist_yl"] = df9["sy_dist"]* 3.26
#filtre taille
df9.dropna(subset=['pl_rade'], inplace=True)
for x in df9.index:
    if df9.loc[x, "pl_rade"] > 10.0:#si la taille de la planete est superieure a 10 terre        df9.drop(x, inplace=True)
        df9.drop(x, inplace=True)
#filtre masse        
df9.dropna(subset=['pl_masse'], inplace=True)
for x in df9.index:
    if df9.loc[x, "pl_masse"] > 10.0:#si la masse de la planete est superieure a 10 terre
        df9.drop(x, inplace=True)
#filtre distance
df9.dropna(subset=['dist_yl'], inplace=True)
for x in df9.index:
    if df9.loc[x, "dist_yl"] > 100.0:#si la distance de la planete est superieure a 100 al
        df9.drop(x, inplace=True)
            
for x in df9.index:
    if df9.loc[x, "discoverymethod"] == "Radial Velocity":
        df9.loc[x,"couleurs"]= "red"
    elif df9.loc[x, "discoverymethod"] == "Transit":
        df9.loc[x,"couleurs"]= "green"
    elif df9.loc[x, "discoverymethod"] == "Imaging":
        df9.loc[x,"couleurs"]= "blue"
couleur = df9["couleurs"]

#print("DF9",df9) 

df10 = df9[['rowid','hostname','pl_name','disc_year','dist_yl',
           'discoverymethod','pl_rade','pl_masse','couleurs']]
#print("DF10",df10)

df11 = df10.groupby(['pl_name'], as_index=False)
#print("DF11max",df11.max())
df12 = df11.max()
couleur = df12["couleurs"]
print(len(df12))
#print("DF12",df12)



ax =df12.plot(kind='scatter', x='disc_year', y='dist_yl', color=couleur, figsize=(12,8))

df12[['disc_year','dist_yl','pl_name']].apply(lambda row: ax.text(*row,ha='right', size=8), axis=1)

plt.title("distance des planetes découverte par année", color='green', fontsize=20)
plt.xlabel("année de découverte", color='purple', fontsize=20)
plt.ylabel("distance en année lumière", color='pink', fontsize=20)
#la terre est le point bleu
plt.plot(2016,0, marker="o", color="blue")
# Creating legend with color box
pop_a = mpatches.Patch(color='red', label='Radial Velocity')
pop_b = mpatches.Patch(color='green', label='Transit')
pop_c = mpatches.Patch(color='blue', label='Imaging')
plt.legend(title="methode",handles=[pop_a,pop_b,pop_c])
# Adding grid to plot
plt.grid()
plt.show()

df13 = df12.sort_values('disc_year', ascending=False)
df13
```
