```python
df["Customer_UpperCase"] = df["Customer Name"].apply(lambda name: name.upper())
```

```python
df
```

```python
df["CustomerLen"] = df.Customer_UpperCase.apply(len)
```

```python
df.head()
```

```python
# "< 0"
# "betwwen 0 - 50"
# "between 51 - 100"
# "> 100"
def calcSales(profit):
    if profit < 0 :
        return "< 0"
    elif profit >=0 and profit <= 50:
        return "between 0 - 50"
    elif profit >=51 and profit <= 100:
        return "between 51 - 100"
    else:
        return "> 100"
```

```python
calcSales(75)
```

```python
df.head()
```

```python
df["Sales Category"] = df.apply(lambda x: calcSales(x["Profit"]), axis =1)
```

```python
df.head()
```

```python
df["Sales Category"].unique()
```

```python
#concatenate 2 columns
def concatenate(col1, col2):
    return str(col1) + "--" + str(col2)
```

```python
df["CombinedColumn"] = df[["Customer Name","City"]].apply(lambda x:
                                                         concatenate(x["Customer Name"],
                                                                     x["City"]),
                                                                     axis = 1)
```

```python
df = pd.DataFrame({"X":np.random.randint(1000, size=1000000),
                   "Y":np.random.randint(1000, size=1000000)})
```

```python
df.head()
```

```python
df.shape
```

```python
def add_squares(df):
    return df.X**2 + df.Y**2
```

```python
%%time
df["Add Square"] = df.apply(add_squares, axis=1)
```
