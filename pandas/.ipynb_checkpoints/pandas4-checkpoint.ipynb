```python
df.query('star_distance.isna()==False')\
.groupby('star_name')['Rayon']\
.agg(['mean','count'])\
.query('count > = 5')\
.sort_values('mean')['mean']\
.plot(kind='barh')
```

```python
df.query('star_distance.isna()==False')\
.groupby('star_name')['Rayon']\
.agg(['mean','count'])\
.query('count > = 5')\
.sort_values('mean')['count']\
.plot(kind='barh')
```

```python
df.query('star_distance.isna()==False')\
.groupby('star_name')['Rayon']\
.agg(['mean','count'])\
.query('count > = 5')\
.sort_values('mean')['count']\
.plot(kind='barh', figsize=(12,5), title="Nombre de planete par etoile")
```

```python
ax = df.query('star_distance.isna()==False')\
.groupby('star_name')['Rayon']\
.agg(['mean','count'])\
.query('count > = 5')\
.sort_values('mean')['count']\
.plot(kind='barh', figsize=(12,5), title="Nombre de planete par etoile")
ax.set_xlabel("Nombre de planete autour de l'etoile")
ax.set_ylabel("Nom de l'etoile")
```
