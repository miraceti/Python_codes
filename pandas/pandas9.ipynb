```python
import pandas as pd
import numpy as np
```

```python
#TCD ou PIVOT TABLE avec PANDAS
```

```python
sample = pd.read_csv("Sample.csv", sep=",")
```

```python
sample.head()
```

```python
sample.columns
```

```python
#creation de la table pivot  "pivot table"
```

```python
pd.pivot_table(
    data=sample,
    index = ["State"]
)
```

```python
pd.pivot_table(sample, index = ["Segment","State"]).head()
```

```python
pd.pivot_table(sample, index = ["Segment","State"], values = ["Sales", "Profit"])
```

```python
pd.pivot_table(sample, index = ["Segment","State"], values = ["Sales", "Profit"],
              aggfunc = np.sum)
```

```python
pd.pivot_table(sample, index = ["Segment","State"], values = ["Sales", "Profit"],
              aggfunc = [np.sum, len])
```

```python
pd.pivot_table(sample, index = ["Category"], values = ["Sales"], columns = ["Segment"],
              aggfunc = np.sum)
```

```python
#missing value or null value
```

```python
pd.pivot_table(sample, index = ["Category"], values = ["Sales"], columns = ["Segment"],
              aggfunc = np.sum, fill_value=0)
```

```python
pd.pivot_table(sample, index = ["Category"], values = ["Sales"], columns = ["Segment"],
              aggfunc = np.sum, fill_value=0, margins=True, margins_name  = "Total")
```

```python
pd.pivot_table(sample, index = ["Category","Segment"], values = ["Sales","Quantity"], 
               aggfunc = {"Sales":np.average, "Quantity":np.sum})
```

```python
pd.pivot_table(sample, index = ["Category","Segment"], values = ["Sales","Quantity"], 
               aggfunc = {"Sales":[np.average,np.sum], "Quantity":np.sum})
```

```python
table = pd.pivot_table(sample, index = ["Category","Segment"], values = ["Sales","Quantity"], 
               aggfunc = {"Sales":[np.average,np.sum], "Quantity":np.sum})
```

```python
table.head()
```

```python
type(table)
```

```python
table.query('Segment == ["Corporate","Home Office"]')
```

```python
pivot = sample.query('Region == "East"')
```

```python
pivot.head()
```

```python
pivot = sample.query('Region == "East"').pivot_table(index="State", 
                                                     columns="Category",
                                                    values  ="Sales",
                                                    aggfunc = "sum",
                                                    margins = True)
```

```python
pivot
```

```python
pivot.sort_values(by = "Furniture")
```

```python
sales = pd.pivot_table(sample, index=["Category"], 
                       values  =["Sales","Quantity"],
                       aggfunc = {"Sales":np.sum,"Quantity":np.sum })
```

```python
sales
```

```python
sales.plot(kind="bar", title="Sales vs Qty")
```

```python
sales = pd.pivot_table(sample, index=["Category"], 
                       values  =["Sales","Profit"],
                       aggfunc = {"Sales":np.sum,"Profit":np.sum })
sales.plot(kind="bar", title="Sales vs profit")
```

```python
sales.plot(kind="line", title="Sales vs profit")
```

```python
sample["Order Date"] = sample["Order Date"].apply(pd.to_datetime)
```

```python
sample.info()
```

```python
pd.pivot_table(sample, index = sample["Order Date"].dt.strftime("%B"),
              values = ["Sales"],
              aggfunc = np.sum).plot(kind='bar')
```

```python
pd.pivot_table(sample, index = sample["Order Date"].dt.strftime("%B"),
              values = ["Sales"],
              aggfunc = np.sum).sort_values(by = "Sales").plot(kind='bar')
```

```python
pd.pivot_table(sample, index = sample["Order Date"].dt.strftime("%B"),
              values = ["Sales"],
              aggfunc = np.sum).sort_values(by = "Sales").plot(kind='line')
```
