```python
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans
    
        
```

```python
def one_step_mult():
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print(a, "x= ", b)
    ans_in = (input("x = "))
    answer = b/a
    
    if string_frac(ans_in)==answer:
        print("Corrrect! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

```python
def one_step_add():
    import random
    a = random.randint(-4,10)
    b = random.randint(2,24)
    print("x + ",a, " = ", b)
    ans = float(input("x = "))
    answer = b-a
    
    if ans==answer:
        print("Corrrect! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
```

# reduce fraction to lowest terms

```python
numerator = 12
denominator = 24
factor = 1

#find the greatest common factoe
for test_factor in range(1, denominator + 1):
    if numerator%test_factor == 0 and denominator%test_factor == 0:
        factor = test_factor

n = int(numerator/factor)
d = int(denominator/factor)

print("original : ", numerator, "/", denominator)
print("reduced", n, "/", d)
```

```python
import math

print(math.sqrt(25))
print(math.sqrt(24))

n = 12

max_factor = 1

upper_limit = math.floor(math.sqrt(n)) + 1

for maybe_factor in range(1, upper_limit):
    if n % (maybe_factor**2) == 0:
        max_factor = maybe_factor
        
        
print("n = ", n)
print("square rooted factor = ", max_factor)
print("square factor = ", max_factor**2)
print("integer = ", n/max_factor**2)
```

# Interactive graphs

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b, zoom):
    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom
    points = 10*(xmax-xmin)
    x = np.linspace(xmin, xmax, points)
    
    plt.axis([xmin,xmax,ymin,ymax])#windows size
    plt.plot([xmin,xmax],[0,0],'k') # black x axis
    plt.plot([0,0],[ymin,ymax],'k') # black y axis
    plt.title('y = mx + b')
    
    plt.plot(x, m*x+b)
    plt.show()
    
#les sliders
interactive_plot = interactive(f, m=(-9, 9), b=(-9, 9), zoom=(1, 100) )
interactive_plot
    
    
```

# Système d'équation 2

```python
from sympy import *

x,y = symbols('x y')

#equation set to 0
first = 2*x + y - 1
second = x - 2*y + 7

#parse the finite set as coordonate pair
solution = linsolve([first, second], (x,y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

print("(", x_solution, ",", y_solution, ")")
```

# factor change from "0 =" to "y =" to graph

```python
from sympy import *
from sympy.plotting import plot

var('x y')

first = 2*x + y - 1
second = x - 2*y + 7

solution = linsolve([first, second], (x,y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]

print("(", x_solution, ",", y_solution, ")")

#sympy syntax for equation equal to 0
y_first = Eq(first, 0)

#sympy solve for y
y1 = solve(y_first,y)

#sympy syntax for equation equal to 0
y_second = Eq(second, 0)

#sympy solve for y
y2 = solve(y_second,y)

#show factored results
print("y = ", y1[0])
print("y = ", y2[0])

#plot solution
x = symbols('x')
plot(y1[0], y2[0], (x, -10, 10))

```
