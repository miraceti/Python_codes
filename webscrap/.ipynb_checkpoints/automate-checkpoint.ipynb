```python
csv_exo3 = csv_exo2[['methode','etoile_distance_pc','etoile_nom',
                     'planete_lettre','planete_nom','planete_rayon_J',
                     'planete_masse_J','periode_jours'
                    ]]
csv_exo3
```

```python
pivot_table = csv_exo3.pivot_table(index='methode', columns='planete_masse_J', values='periode_jours',
                    aggfunc='sum')
pivot_table
```

```python
#export pivot_table to excel
pivot_table.to_excel('pivot_table.xlsx','data_methode', startrow=4)
```

```python
#create a bar chart with openpyxl
import openpyxl
```

```python
from openpyxl import load_workbook
from openpyxl.chart import BarChart, Reference  

wb = load_workbook('pivot_table.xlsx')
sheet = wb['data_methode']

min_column = wb.active.min_column
max_column = wb.active.max_column
min_row = wb.active.min_row
max_row = wb.active.max_row
print("columns MIN: ", min_column, " - MAX: ", max_column )
print("rows    MIN: ", min_row, " - MAX: ", max_row )
```

```python
barchart = BarChart()

data = Reference(sheet,min_col=min_column+1,max_col=max_column,min_row=min_row,max_row = max_row)

categories=Reference(sheet,min_col=min_column,max_col=min_column,min_row=min_row+1,max_row=max_row)

barchart.add_data(data, titles_from_data = True)
barchart.set_categories(categories)

sheet.add_chart(barchart, "B30")

barchart.title = "exoplanet methods"
barchart.style = 5
wb.save("barchart.xlsx")
```
