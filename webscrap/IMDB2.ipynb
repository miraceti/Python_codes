```python
from bs4 import BeautifulSoup
import requests
import openpyxl

excel = openpyxl.Workbook()
print(excel.sheetnames)

sheet = excel.active
sheet.title = ' Top rated movies'
print(excel.sheetnames)

#headings columns
sheet.append(['Movie num','Movie name'])
```

```python
try:
    source = requests.get(
        'https://fr.wikipedia.org/wiki/Liste_de_films_fran%C3%A7ais_sortis_en_2022'
                        )
    source.raise_for_status()
    
    soup = BeautifulSoup(source.text, 'html.parser')
    #print(soup)
    
    tablefilm = soup.find('table', class_="wikitable sortable")
    #print(tablefilm)   
    
    #movies = soup.find('tbody', class_="nomclass")
    movies = soup.find('tbody')#the first match et pas de class
    #print(movies)
    
    #tous les TR tag
    movies_list = soup.find('tbody').find_all('tr')
    print("\n***************LEn MOVIES LIST*******************************\n")
    #print(len(movies_list))
    print("\n***************MOVIES LIST*******************************\n")
    #print(movies_list)
    
    nblignes = len(movies_list)
    
    print("\n\n\n***************MOVIES LIST 10*******************************\n")
    print(movies_list[10])
    film = movies_list[10]
    print("\n\n\n***************MOVIES LIST*******************************\n")
    #for movie in movies_list:
        #print(movie)
    print("\n\n\n***************MOVIE bloc*******************************\n")    
    name = film.find('td').a
    print("name : ", name)
    
    print("\n\n\n***************MOVIES name href*******************************\n")
    name = film.find('td').a
    print("name : ", name)
    
    print("\n\n\n***************MOVIES name text*******************************\n")
    name_text = film.find('td').a.text
    print("name_text : ", name_text)
    
    print("\n\n\n***************MOVIE  text*******************************\n")
    nametext = film.find('td').text
    print("nametext : ", nametext)
    
    print("\n\n\n***************MOVIE  gettext*******************************\n")
    get_text = film.find('td').get_text(strip=True)
    print("get_text : ", get_text)
    
    #split
    get_text_split = film.find('td').get_text(strip=True).split('.')
    print("get_text_split : ", get_text_split)
    
    get_text_split0 = film.find('td').get_text(strip=True).split('.')[0]
    print("get_text_split0 : ", get_text_split0)
    
    #si year= (1994)
    #get_year = film.find('td').span.text.strip('()')
    #print("get_year : ", get_year) ==> 1994
    
    #"on boucle sur tutes les lignes"
    
    for n in range(1,nblignes):
        print(n)
        film_n = movies_list[n]
        #print(film_n)
        namen = film_n.find('td').a
        print("name : ", namen)
        if namen == None:
            print(n, "vide")
        else:
            name_text_n = film_n.find('td').a.text
            print("name_text : ", name_text_n)
            sheet.append([n,name_text_n])
    
except Exception as e:
    print(e)
    
excel.save('imdb2.xlsx')
```
